```python
import pyblp
import numpy as np
import pandas as pd
import copy
import statsmodels.formula.api as smf
from matplotlib import pyplot as plt

pyblp.options.digits = 4
pyblp.options.verbose = False
pyblp.__version__
```

```python
# Read in data
agent_data = pd.read_csv('./agent_data.csv', index_col=[0])
product_data = pd.read_csv('./product_data.csv', index_col=[0])
```

<h3> Question 1(a): OLS Model </h3>

```python
# Setup data
product_data_ols = copy.copy(product_data)
product_data_ols['share_outside'] = 1 - (
    product_data_ols
    .groupby('market_ids', as_index=False)['shares']
    .transform(sum)
)
product_data_ols['diff_log_shares'] = (
    np.log(product_data_ols['shares'])
    - np.log(product_data_ols['share_outside'])
)
```

```python
# Do OLS and print result
model_ols = smf.ols(
    formula='diff_log_shares ~ prices + sugar',
    data=product_data_ols
)
result_ols = model_ols.fit()
print(result_ols.summary())
```

<h3> Question 1(b): 2SLS model </h3>

```python
# Do 2SLS and print
formulation_2sls = pyblp.Formulation('1 + prices + sugar')
problem_2sls = pyblp.Problem(formulation_2sls, product_data)
results_2sls = problem_2sls.solve()
print(results_2sls)
```

<h3> Question 1(c): Graph elasticities </h3>

```python
elasticities = results_2sls.compute_elasticities()
mask = (product_data['market_ids'] == 'C01Q1')
own_elasticities_C01Q1 = np.diag(elasticities[mask])
plt.scatter(
    product_data.loc[mask, 'prices'],
    own_elasticities_C01Q1
)
plt.xlabel('prices')
plt.ylabel('elasticities')
plt.show()
```

*Comment*: there is a negative linear relationship between prices and the own price elasticity. Of course, as we discussed in the lecture, this is not a `finding' but a mechanical consequence of our restrictive homogenous-consumer logit model.

<h3> Question 2(a): BLP Model</h3>


```python
# Setup the 'formulations'
X1_formulation = pyblp.Formulation('1 + prices')
X2_formulation = pyblp.Formulation('1 + prices + sugar')
product_formulations = (X1_formulation, X2_formulation)
agent_formulation = pyblp.Formulation('0 + income')
```

```python
# Setup the 'problem' class
nevo_problem = pyblp.Problem(
    product_formulations,
    product_data,
    agent_formulation,
    agent_data
)
```

```python
# Configure some settings for the solver
initial_sigma = np.diag([0, 1.0, 0])
initial_pi = np.array([[5.4819], [5.8935], [-0.2506]])
tighter_bfgs = pyblp.Optimization('bfgs', {'gtol': 1e-5})
```

```python
# Do BLP
nevo_results = nevo_problem.solve(
    initial_sigma,
    initial_pi,
    optimization=tighter_bfgs,
    method='1s'
)
print(nevo_results)
```

<h3> Question 2(b): Own-price Elasticities</h3>

```python
nevo_elasticities = nevo_results.compute_elasticities()
mask = (product_data['market_ids'] == 'C01Q1')
nevo_own_elasticities_C01Q1 = np.diag(nevo_elasticities[mask])
plt.scatter(
    product_data.loc[mask, 'prices'],
    nevo_own_elasticities_C01Q1
)
plt.xlabel('prices')
plt.ylabel('elasticities')
plt.show()
```

*Comment:* Unlike in Part 1(c) there is no longer a negative relationship between the elasticities and prices. This is because we are now allowing for heterogeneous consumer tastes and so the data are now flexibly determining this relationship.

From the graph above it looks like there might in fact be a positive correlation between prices and elastitices. I would caution against concluding this without more model robustness tests because we are still using a very restrictive model to estimate demand. We are using a restrictive model because it is easy to compute in the assignment but in practice there should probably be more product characteristics and consumer demographics (e.g. the 'mushy' variable from Nevo (2000)).
